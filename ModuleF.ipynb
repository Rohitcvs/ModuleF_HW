Section 1: Numerical Differentiation Problem Statement

Numerical differentiation is a computational technique for estimating derivatives when an explicit analytical form of the function's derivative is unavailable or difficult to compute. This is especially useful in real-world applications where data is discrete, such as in engineering, finance, and physics.

A common application of numerical differentiation is in scientific computing, where measurements from experiments are used to determine rates of change. For example, in physics, numerical differentiation helps estimate velocity and acceleration from position data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a sample function
def f(x):
    return np.sin(x)

# Compute the analytical derivative
def df_analytical(x):
    return np.cos(x)

# Plot the function and its derivative
x_vals = np.linspace(0, 2*np.pi, 100)
y_vals = f(x_vals)
dy_vals = df_analytical(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = sin(x)")
plt.plot(x_vals, dy_vals, label="df/dx = cos(x)", linestyle='dashed')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Function and its Analytical Derivative")
plt.show()
```

Section 2: Finite Difference Approximating Derivatives

Finite difference methods approximate derivatives using discrete points. The most common formulas are:
1. Forward Difference: f'(x) ~ f(x+h)-f(x)/h
2. Backward Difference: f'(x) ~ f(x) - f(x-h)/h
3. Central Difference: f'(x) ~ f(x+h) - f(x-h)/2h (more accurate)

The central difference method is generally preferred because it reduces error compared to forward and backward differences.

```python
# Finite difference approximation
def forward_difference(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

def backward_difference(f, x, h=1e-5):
    return (f(x) - f(x - h)) / h

def central_difference(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2*h)

# Compute derivatives at a point
x0 = np.pi / 4  # 45 degrees
true_derivative = df_analytical(x0)

forward_approx = forward_difference(f, x0)
backward_approx = backward_difference(f, x0)
central_approx = central_difference(f, x0)

print(f"True derivative at x={x0}: {true_derivative:.5f}")
print(f"Forward Difference Approximation: {forward_approx:.5f}")
print(f"Backward Difference Approximation: {backward_approx:.5f}")
print(f"Central Difference Approximation: {central_approx:.5f}")
```

Section 3: Approximation Higher-Order Derivatives

Higher-order derivatives describe the rate of change of lower-order derivatives. For example, in physics:
1. The first derivative of position is velocity.
2. The second derivative of position is acceleration.

For a function f(x), the second derivative can be approximated using the second-order central difference formula:
    f''(x) ~ f(x+h) - 2f(x) + f(x-h) / h^2
This approach can be extended to compute even higher-order derivatives, but the error increases with each order.


```python
# Approximate second derivative using finite difference
def second_derivative(f, x, h=1e-5):
    return (f(x + h) - 2*f(x) + f(x - h)) / h**2

# Compute second derivative at a point
second_deriv_approx = second_derivative(f, x0)
true_second_deriv = -np.sin(x0)  # Analytical second derivative of sin(x)

print(f"True second derivative at x={x0}: {true_second_deriv:.5f}")
print(f"Approximated Second Derivative: {second_deriv_approx:.5f}")
```
